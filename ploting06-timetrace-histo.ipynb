```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#For making plot dimensions in cm
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=[9*cm,6*cm]) 

#To set the fontsize of various variables
import  matplotlib as mpl
SMALL_SIZE=12
MEDIUM_SIZE=14
BIGGER_SIZE=16
mpl.rc('font', size=SMALL_SIZE)          # controls default text sizes
mpl.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
mpl.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
mpl.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
mpl.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
mpl.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
mpl.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
a10, b10, c10, d10, e10  = np.loadtxt('COLVAR-PLUM-DR10.00',comments=['@', '#'], unpack=True)
a12, b12, c12, d12, e12  = np.loadtxt('COLVAR-PLUM-DR12.00',comments=['@', '#'], unpack=True)
a15, b15, c15, d15, e15  = np.loadtxt('COLVAR-PLUM-DR15.00',comments=['@', '#'], unpack=True)
a18, b18, c18, d18, e18  = np.loadtxt('COLVAR-PLUM-DR18.00',comments=['@', '#'], unpack=True)


cm = 1/2.54  # centimeters in inches
plt.figure(figsize=[9*cm,6*cm]) 
#fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6,1, figsize=(8*cm,40*cm))

left, width = 0.18, 0.60
bottom, height = 0.20, 0.70
spacing = 0.005

rect_scatter = [left, bottom, width, height]
#rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

#plotting data into different boxes
ax1_scatter = plt.axes(rect_scatter)
ax1_scatter.tick_params(direction='out', top=False, right=False)
ax1_scatter = plt.plot(a10/1000, b10*57.2958, "-r", linewidth=0.5, label="10.00")

#plt.plot(y['time']/1000, ang_y, "-b", linewidth=0.5, alpha=0.8 )

ax1_scatter = plt.plot(a12/1000, b12*57.2958, "-g", linewidth=0.5, label="12.00")

ax1_scatter = plt.plot(a15/1000, b15*57.2958, "-b", linewidth=0.5, label="15.00")
ax1_scatter = plt.plot(a18/1000, b18*57.2958, "-", color='magenta', linewidth=0.5, label="18.00")
ax1_scatter= plt.axhline(y=60, linestyle='--', color='black', linewidth=1, alpha=0.5)


ax1_scatter= plt.xlim([0, 76])                                  #xrange
ax1_scatter = plt.ylim([30, 101])     

ax1_scatter =plt.xticks(np.arange(0, 76, 15))        #ticks
ax1_scatter =plt.yticks(np.arange(30, 101, 15))        #ticks


from matplotlib.ticker import MultipleLocator
ax1_scatter= plt.gca().yaxis.set_minor_locator(MultipleLocator(5))
ax1_scatter= plt.gca().xaxis.set_minor_locator(MultipleLocator(5))

ax1_scatter = plt.xlabel('time (ns)')                      #xlabel
ax1_scatter = plt.ylabel('${\mathrm{\\theta_H}}$ ($^{\circ}$)')   

#for x axis distribution plot
#ax_histx = plt.axes(rect_histx)
#ax_histx.tick_params(direction='in', labelbottom=False)

#for y axis distribution plot
ax1_histy = plt.axes(rect_histy)
ax1_histy.tick_params(direction='in', labelleft=False)
ax1_histy.tick_params(direction='in', labelbottom=False)
ax1_histy =plt.hist(b10*57.2958, bins=100, histtype='step', color='red', label='10.00', align='mid',  linewidth=1.,  density=True, orientation='horizontal')
ax1_histy =plt.hist(b12*57.2958, bins=100, histtype='step', color='green', label='12.00', align='mid', linewidth=1., density=True, orientation='horizontal')
ax1_histy =plt.hist(b15*57.2958, bins=100, histtype='step', color='blue', label='15.00',  align='mid', linewidth=1., density=True, orientation='horizontal')
ax1_histy =plt.hist(b18*57.2958, bins=100, histtype='step', color='magenta', label='18.00', align='mid', linewidth=1., density=True, orientation='horizontal')
ax1_histy =plt.axhline(y=60, linestyle='--', color='black', lw=1.0, alpha=0.5)

ax1_histy= plt.xlim([0, 0.15])
ax1_histy = plt.ylim([30, 101])

ax1_histy =plt.yticks(np.arange(30, 101, 15) )        #ticks


plt.savefig('image-Mangle-tass.png', dpi=300, bbox_inches='tight',pad_inches = 0.1) 
```
