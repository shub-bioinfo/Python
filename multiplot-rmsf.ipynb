```python
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt('rmsfApoA1.xvg',comments=['@', '#'], unpack=True)
```

```python
plt.figure(figsize=[3.5,2.7])   #Plot Size

#linewidth, alpha for occupancy of line color 
plt.plot(x, y*10, '-r', label='A1', linewidth=1.0, alpha=0.5)        #Line plot

plt.xlabel('residue number', fontsize=14)           #xlabel
plt.ylabel(r'rmsf (${\mathrm{\AA}}$)', fontsize=14)                     #ylabel

plt.xlim(536, 859)                                  #xrange
plt.ylim(0, 4)                                      #yrange

plt.xticks(np.arange(536, 859, 50 ), fontsize=14)         #ticks
plt.yticks(np.arange(0,4,1), fontsize=14)

#To hide the top and right axis of the plot
ax=plt.gca()
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

#To add colored background slabs in the plot
ax.axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
ax.axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')

#To add legend, position, remove legend line, remove frame, labelcolor as line color and fontsize.
ax.legend(loc='upper right', handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=18)

#to remove whitespace, and set dpi
plt.savefig('image1.png', dpi = 300, bbox_inches='tight',pad_inches = 0) 
#plt.show() 
```
