```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

t, angle, helicity, interaction, bias = np.loadtxt('COLVARApo-A3',comments=['@', '#'], unpack=True)

t1, dis1 = np.loadtxt('distPHE786-820-apo.xvg',comments=['@', '#'], unpack=True)
```

```python
#plt.rc('font', family='Helvetica')
cm = 1/2.54  # centimeters in inches


#plt.figure(figsize=[3.5,2.7])   #Plot Size
#To create three suplots
fig, ax = plt.subplots(2, 2, figsize=(19*cm,16*cm))
fig.subplots_adjust(hspace=0.25, wspace=0.35)

#linewidth, alpha for occupancy of line color 
ax[0, 0].plot(t/1000, angle*57.2958, '-r', linewidth=0.8, alpha=0.8)        #Line plot
ax[0, 0].axhline(y=60, linestyle='--', color='black')

ax[0, 1].plot(t/1000, helicity, '-g', linewidth=0.8, alpha=0.8)        #Line plot


ax[1, 0].plot(t/1000, interaction, '-b', linewidth=0.8, alpha=0.8)        #Line plot

ax[1, 1].plot(t1/1000, dis1*10, 'magenta', linewidth=0.8, alpha=0.8)        #Line plot

ax[0, 0].set_ylabel('${\mathrm{\\theta_H}}$ ($^{\circ}$)', fontsize=16)   
ax[0, 1].set_ylabel('helicity', fontsize=16)                               #ylabel
ax[1, 0].set_ylabel('interactions', fontsize=16)                             #ylabel
ax[1, 1].set_ylabel('distance (${\AA}$)', fontsize=16)                             #ylabel


ax[0, 0].set_xlabel('', fontsize=16)           #xlabel
ax[0, 1].set_xlabel('', fontsize=16)           #xlabel
ax[1, 0].set_xlabel('time (ns)', fontsize=16)           #xlabel
ax[1, 1].set_xlabel('time (ns)', fontsize=16)           #xlabel


ax[0, 0].set_xlim([0, 301])                                  #xrange
ax[0, 1].set_xlim([0, 301])                                  #xrange
ax[1, 0].set_xlim([0, 301])                                  #xrange
ax[1, 1].set_xlim([0, 301]) 

ax[0, 0].set_ylim([20, 121])                                      #yrange
ax[0, 1].set_ylim([0, 8])                                      #yrange
ax[1, 0].set_ylim([0, 15])                                      #yrange
ax[1, 1].set_ylim([0, 18])                                      #yrange


ax[0, 0].set_xticks(np.arange(0, 301, 100), fontsize=14)         #ticks
ax[0, 0].set_yticks(np.arange(20, 121, 20), fontsize=14)
ax[0, 0].text(-100, 125, 'A', fontsize=18, weight='bold')
ax[0, 0].text(40, 40, 'inward', fontsize=14)
ax[0, 0].text(200, 90, 'outward', fontsize=14)

ax[0, 1].set_xticks(np.arange(0, 301, 100), fontsize=14)         #ticks
ax[0, 1].set_yticks(np.arange(0, 9, 1), fontsize=14)
ax[0, 1].text(-80, 8.4, 'B', fontsize=18, weight='bold')

ax[1, 0].set_xticks(np.arange(0, 301, 100), fontsize=14)         #ticks
ax[1, 0].set_yticks(np.arange(0, 16, 2), fontsize=14)
ax[1, 0].text(-100, 15.75, 'C', fontsize=18, weight='bold')

ax[1, 1].set_xticks(np.arange(0, 301, 100), fontsize=14)         #ticks
ax[1, 1].set_yticks(np.arange(0, 16, 5), fontsize=14)
ax[1, 1].text(-80, 18.9, 'D', fontsize=18, weight='bold')
ax[1, 1].text(100, 16, 'Phe786-Phe820', fontsize=14, color='magenta')

from matplotlib.ticker import MultipleLocator
ax[0, 0].xaxis.set_minor_locator(MultipleLocator(50))
ax[0, 1].xaxis.set_minor_locator(MultipleLocator(50))
ax[0, 1].xaxis.set_minor_locator(MultipleLocator(50))
ax[1, 1].xaxis.set_minor_locator(MultipleLocator(50))

#ax[0, 0].yaxis.set_minor_locator(MultipleLocator(10))


#To hide the top and right axis of the plot
#ax=ax[0, 0].gca()
# Hide the right and top spines
ax[0, 0].spines['right'].set_visible(False)
ax[0, 0].spines['top'].set_visible(False)
ax[0, 1].spines['right'].set_visible(False)
ax[0, 1].spines['top'].set_visible(False)
ax[1, 0].spines['right'].set_visible(False)
ax[1, 0].spines['top'].set_visible(False)
ax[1, 1].spines['right'].set_visible(False)
ax[1, 1].spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax[0, 0].yaxis.set_ticks_position('left')
ax[0, 0].xaxis.set_ticks_position('bottom')
ax[0, 1].yaxis.set_ticks_position('left')
ax[0, 1].xaxis.set_ticks_position('bottom')
ax[1, 0].yaxis.set_ticks_position('left')
ax[1, 0].xaxis.set_ticks_position('bottom')
ax[1, 1].yaxis.set_ticks_position('left')
ax[1, 1].xaxis.set_ticks_position('bottom')

# #To add colored background slabs in the plot
# ax[0, 0].axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
# ax[0, 0].axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')
# ax[0, 1].axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
# ax[0, 1].axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')
# ax[0, 1].axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
# ax[0, 1].axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')

#To add legend, position, remove legend line, remove frame, labelcolor as line color and fontsize.
#bbox_to_anchor: to position the legend location
# ax[0, 0].legend(loc='upper right', bbox_to_anchor=(1.05, 1.1), handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=16)
# ax[0, 1].legend(loc='upper right', bbox_to_anchor=(1.05, 1.1), handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=16)
# ax[0, 1].legend(loc='upper right', bbox_to_anchor=(1.05, 1.1), handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=16)


#To manage the horizontal spacing between legends 
# plt.rcParams["legend.labelspacing"]=0.1

#to remove whitespace, and set dpi
fig.savefig('figure-CV-pi-Apo.pdf', dpi = 300, bbox_inches='tight',pad_inches = 0) 
plt.show() 
plt.clf()
plt.close()
```
