```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset=pd.read_table('COLVAR_DISTx', delim_whitespace=True ,skiprows=13, header=None) 

dataset[[1, 2, 3, 4]].corr()
```

```python
#To remove the first column of time inplace in the dataset
dataset.drop(columns=[0], axis=1, inplace=True)
dataset

```

```python
corr=dataset.corr()
#mask = np.triu(np.ones_like(corr, dtype=bool))
plt.subplots(figsize=(25,18))
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True)
plt.savefig('image.png', dpi = 300, bbox_inches='tight',pad_inches = 0) 
plt.show()
```
