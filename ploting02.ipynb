```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

t, angle, helicity, interaction, bias = np.loadtxt('COLVAREvo-B5',comments=['@', '#'], unpack=True)

t
```

```python
#plt.rc('font', family='Helvetica')
cm = 1/2.54  # centimeters in inches


#plt.figure(figsize=[3.5,2.7])   #Plot Size
#To create three suplots
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(12*cm,30*cm))

#linewidth, alpha for occupancy of line color 
ax1.plot(t/1000, angle*57.2958, '-r', label='A1', linewidth=1.0, alpha=0.8)        #Line plot
ax1.axhline(y=60, linestyle='--', color='black')

ax2.plot(t/1000, helicity, '-g', label='B1', linewidth=1.0, alpha=0.8)        #Line plot


ax3.plot(t/1000, interaction, '-b', label='B2', linewidth=1.0, alpha=0.8)        #Line plot


ax1.set_ylabel('${\mathrm{\\theta_H}}$ ($^{\circ}$)', fontsize=16)   
#ax1.set_ylabel('angle (${\theta}{\mathrm_{H}}$)', fontsize=16)                     #ylabel
ax2.set_ylabel('helicity', fontsize=16)                               #ylabel
ax3.set_ylabel('interactions', fontsize=16)                             #ylabel

ax1.set_xlabel('', fontsize=16)           #xlabel
ax2.set_xlabel('', fontsize=16)           #xlabel
ax3.set_xlabel('time (ns)', fontsize=16)           #xlabel

ax1.set_xlim(0, 251)                                  #xrange
ax2.set_xlim(0, 251)                                  #xrange
ax3.set_xlim(0, 251)                                  #xrange

ax1.set_ylim(20, 121)                                      #yrange
ax2.set_ylim(0, 8)                                      #yrange
ax3.set_ylim(0, 16)                                      #yrange

ax1.set_xticklabels(np.arange(0, 251, 50 ), fontsize=14)         #ticks
ax1.set_yticklabels(np.arange(20, 121, 20), fontsize=14)
ax1.text(-60, 120, 'A', fontsize=18, weight='bold')
ax1.text(120, 30, 'inward', fontsize=16)
ax1.text(120, 100, 'outward', fontsize=16)

ax2.set_xticklabels(np.arange(0, 301, 50  ), fontsize=14)         #ticks
ax2.set_yticklabels(np.arange(0, 8, 1), fontsize=14)
ax2.text(-60, 8, 'B', fontsize=18, weight='bold')

ax3.set_xticklabels(np.arange(0, 301, 50 ), fontsize=14)         #ticks
ax3.set_yticklabels(np.arange(0, 15, 1), fontsize=14)
ax3.text(-60, 15, 'C', fontsize=18, weight='bold')

from matplotlib.ticker import MultipleLocator
ax1.xaxis.set_minor_locator(MultipleLocator(50))
ax2.xaxis.set_minor_locator(MultipleLocator(50))
ax3.xaxis.set_minor_locator(MultipleLocator(50))

#ax1.yaxis.set_minor_locator(MultipleLocator(10))


#To hide the top and right axis of the plot
#ax=ax1.gca()
# Hide the right and top spines
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax1.yaxis.set_ticks_position('left')
ax1.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
ax2.xaxis.set_ticks_position('bottom')
ax3.yaxis.set_ticks_position('left')
ax3.xaxis.set_ticks_position('bottom')

# #To add colored background slabs in the plot
# ax1.axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
# ax1.axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')
# ax2.axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
# ax2.axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')
# ax3.axvspan(659, 679, ymin=0, ymax=4, alpha=0.1, color='red')
# ax3.axvspan(772, 797, ymin=0, ymax=4, alpha=0.1, color='blue')

#To add legend, position, remove legend line, remove frame, labelcolor as line color and fontsize.
#bbox_to_anchor: to position the legend location
# ax1.legend(loc='upper right', bbox_to_anchor=(1.05, 1.1), handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=16)
# ax2.legend(loc='upper right', bbox_to_anchor=(1.05, 1.1), handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=16)
# ax3.legend(loc='upper right', bbox_to_anchor=(1.05, 1.1), handlelength=0, handletextpad=0, frameon=False, labelcolor='linecolor', fontsize=16)


#To manage the horizontal spacing between legends 
# plt.rcParams["legend.labelspacing"]=0.1

#to remove whitespace, and set dpi
fig.savefig('figure-CV-Evo.png', dpi = 300, bbox_inches='tight',pad_inches = 0) 
#plt.show() 
```
