# This is to plot the Apo data

```python
x=pd.read_table('ApoSalt-format.dat', delimiter=' ')
#df =pd.DataFrame(x, index=x['salt'], columns=['A1','A11','A2','A21','A3','A31'])

#Pivoting the data for the heatmap plot
df =pd.pivot_table(x, index=['Salt-bridge'], columns=['Simulation'], values=['Occupancy'])

#Using this style is removing the additional occupancy part in xlabel but it is changing the order of xtics
#df=x.pivot_table('Occupancy', index=['Salt-bridge'], columns=['Simulation'])

x.head()
df
#x_axis_label=['A1', 'A1', 'A2' , 'A2' , 'A3', 'A3']
#xticks=['A1$^{\prime}$','A1$^{\prime\prime}$','A2$^{\prime}$','A2$^{\prime\prime}$','A3$^{\prime}$','A3$^{\primme\prime}$']

#To define the size of plot and top and bottom margins
fig,ax=plt.subplots(1,1, figsize=(7,3.))
plt.gcf().subplots_adjust(left=0.45)
plt.gcf().subplots_adjust(bottom=0.2)

#Annotate to write the values inside the boxes, and its fontsize. Cbar max and min values
s=sns.heatmap(df, annot=True, annot_kws={"size":16}, vmax=90, vmin=0)

#Cbar label and and fontsizee
ax.figure.axes[-1].set_ylabel('Occupancy %', size=16)

#Cbar ticks fontsize
cbar=ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=14)

#Maintaining the font size of the x and y labels and ticks
s.set_xlabel('', fontsize=20)
s.set_ylabel('', fontsize=20)

s.set_xticklabels(s.get_xmajorticklabels(), fontsize=16)
s.set_yticklabels(s.get_ymajorticklabels(), fontsize=16, rotation=0)

plt.savefig('ApoSalt-Map.png', dpi=600)
```

# This is for the Avanafil Run data

```python
x=open('Avansalt-format.dat','w')
x.write("Salt-bridge Simulation Occupancy \n")

with open('Avansalt-form.dat', 'r') as text:
    for line in text:
        f=line.strip().split('&')
        
        #Creating the simulaiton run list
        y=['1V1$^{\prime}$','2V1$^{\prime\prime}$','3V2$^{\prime}$','4V2$^{\prime\prime}$']
        
        #Looping over words of line
        for i in range(1, len(f)):
            #print(f[0], y[i-1], f[i])
            x.write("{} {} {} \n".format(f[0], y[i-1], f[i]))
# line = "{} is {}} ag {}\n".format(city_name, country_name, rating)
        
x.close()
```

```python
file=pd.read_table('Avansalt-format.dat', delim_whitespace=True)
#df =pd.DataFrame(x, index=x['salt'], columns=['A1','A11','A2','A21','A3','A31'])
df =pd.pivot_table(file, index=['Salt-bridge'], columns=['Simulation'], values=['Occupancy'])
#df=x.pivot_table('Occupancy', index=['Salt-bridge'], columns=['Simulation'])

#x.head()
file
#x_axis_label=['A1', 'A1', 'A2' , 'A2' , 'A3', 'A3']
#xticks=['A1$^{\prime}$','A1$^{\prime\prime}$','A2$^{\prime}$','A2$^{\prime\prime}$','A3$^{\prime}$','A3$^{\primme\prime}$']

fig,ax=plt.subplots(1,1, figsize=(7,3))
plt.gcf().subplots_adjust(left=0.45)
plt.gcf().subplots_adjust(bottom=0.2)
s=sns.heatmap(df, annot=True, annot_kws={"size":16}, vmax=90, vmin=0)

ax.figure.axes[-1].set_ylabel('Occupancy %', size=16)

cbar=ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=14)

s.set_xlabel('', fontsize=20)
s.set_ylabel('', fontsize=20)

s.set_xticklabels(s.get_xmajorticklabels(), fontsize=16)
s.set_yticklabels(s.get_ymajorticklabels(), fontsize=16, rotation=0)
#s.set_xticks(range(len(df)))
#s.set_xtickslabels('a', 'c', 'c', 'c','d','g')
#s.set_xticklabels(['A1$^{\prime}$','A1$^{\prime\prime}$','A2$^{\prime}$','A2$^{\prime\prime}$','A3$^{\prime}$','A3$^{\primme\prime}$'])

plt.savefig('AvanSalt-Map.png', dpi=600)
```
