```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#For making plot dimensions in cm
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=[9*cm,6*cm]) 

#To set the fontsize of various variables
import  matplotlib as mpl
SMALL_SIZE=8
MEDIUM_SIZE=10
BIGGER_SIZE=16
mpl.rc('font', size=SMALL_SIZE)          # controls default text sizes
mpl.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
mpl.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
mpl.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
mpl.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
mpl.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
mpl.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
aH09=pd.read_table('rmsdHL09.00.xvg', names=['time', 'rmsd'], comment='@', delim_whitespace=True ,skiprows=18, header=None) 
aM09=pd.read_table('rmsdML09.00.xvg', names=['time', 'rmsd'], comment='@', delim_whitespace=True ,skiprows=18, header=None) 
aP09=pd.read_table('rmsdPro09.00.xvg', names=['time', 'rmsd'], comment='@', delim_whitespace=True ,skiprows=18, header=None) 


cm = 1/2.54  # centimeters in inches
plt.figure(figsize=[7*cm,4*cm]) 
#fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6,1, figsize=(8*cm,40*cm))

left, width = 0.18, 0.60
bottom, height = 0.20, 0.70
spacing = 0.005

rect_scatter = [left, bottom, width, height]
#rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

#plotting data into different boxes
ax1_scatter = plt.axes(rect_scatter)
ax1_scatter.tick_params(direction='out', top=False, right=False)

#10
ax1_scatter = plt.plot(aH09['time'], aH09['rmsd']*10, "-r", label="A2", linewidth=0.5, alpha=0.2 )
avgH09=aH09.iloc[:,1].rolling(window=200).mean()
ax1_scatter =plt.plot(aH09['time'], avgH09*10, "-r", linewidth=0.5, alpha=0.8 )
#plt.plot(y['time']/1000, ang_y, "-b", linewidth=0.5, alpha=0.8 

#12
ax1_scatter = plt.plot(aM09['time'], aM09['rmsd']*10, "-g", label="A2", linewidth=0.5, alpha=0.2 )
avgM09=aM09.iloc[:,1].rolling(window=200).mean()
ax1_scatter =plt.plot(aM09['time'], avgM09*10, "-g", linewidth=0.5, alpha=0.8 )

#15
ax1_scatter = plt.plot(aP09['time'], aP09['rmsd']*10, "-b", label="A2", linewidth=0.5, alpha=0.2 )
avgP09=aP09.iloc[:,1].rolling(window=200).mean()
ax1_scatter =plt.plot(aP09['time'], avgP09*10, "-b", linewidth=0.5, alpha=0.8 )


ax1_scatter= plt.xlim([0, 75])                                  #xrange
ax1_scatter = plt.ylim([0, 13])     

ax1_scatter =plt.xticks(np.arange(0, 76, 15))        #ticks
ax1_scatter =plt.yticks(np.arange(0, 14, 3))        #ticks



from matplotlib.ticker import MultipleLocator
ax1_scatter= plt.gca().yaxis.set_minor_locator(MultipleLocator(1))
ax1_scatter= plt.gca().xaxis.set_minor_locator(MultipleLocator(5))

# ax1_scatter = plt.xlabel('time (ns)')                      #xlabel
ax1_scatter = plt.ylabel('RMSD ($\mathrm {\AA}$)')   

#for x axis distribution plot
#ax_histx = plt.axes(rect_histx)
#ax_histx.tick_params(direction='in', labelbottom=False)


#for y axis distribution plot
ax1_histy = plt.axes(rect_histy)
ax1_histy.tick_params(direction='in', left=False, right=True, bottom=False, labelleft=False)
ax1_histy.tick_params(direction='in', labelbottom=False)
ax1_histy =plt.hist(aH09['rmsd']*10, bins=100, histtype='step', color='red', label='10.00',  linewidth=0.7,  density=True, orientation='horizontal')
ax1_histy =plt.hist(aM09['rmsd']*10, bins=100, histtype='step', color='green', label='12.00', linewidth=0.7, density=True, orientation='horizontal')
ax1_histy =plt.hist(aP09['rmsd']*10, bins=100, histtype='step', color='blue', label='15.00',  linewidth=0.7, density=True, orientation='horizontal')
# ax1_histy =plt.hist(a18['rmsd'], bins=100, histtype='step', color='magenta', label='18.00', linewidth=1., density=True, orientation='horizontal')
#ax1_histy =plt.axhline(y=60, linestyle='--', color='black', lw=1.0, alpha=0.5)
ax1_histy =plt.text(0.9, 11, '9', fontsize=12, verticalalignment='center')
#ax1_histy =plt.text(0.038, 35, 'inward', fontsize=10)


ax1_histy= plt.xlim([0, 01.40])
ax1_histy = plt.ylim([0, 13])

ax1_histy =plt.yticks(np.arange(0, 14 ,3) )        #ticks

plt.savefig('image-rmsd09.png', dpi=300, bbox_inches='tight',pad_inches = 0.1) 
```
