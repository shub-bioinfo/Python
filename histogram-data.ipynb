```python
import numpy as np
import matplotlib.pyplot as plt



import  matplotlib as mpl
SMALL_SIZE=12
MEDIUM_SIZE=14
BIGGER_SIZE=28
mpl.rc('font', size=SMALL_SIZE)          # controls default text sizes
mpl.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
mpl.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
mpl.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
mpl.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
mpl.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
mpl.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

```

```python
a =np.loadtxt('residence_timeTASS6-8-12.dat', usecols=[1])
print(a)

b =np.loadtxt('residence_timeUS6-8-12.dat', usecols=[1])
print(b)
!pwd
```

```python
plt.plot(b)
```

```python
cm=1/2.54
plt.figure(figsize=[9*cm, 5*cm]) 

plt.hist(a, bins=100, range=(0,2000), label='TASS 6.00', facecolor='r', alpha=0.6)
plt.hist(b, bins=100, range=(0,2000), label='US 6.00', facecolor='b', alpha=0.5)
plt.xlabel('Time (ps)')
plt.ylabel('Water molecules')

plt.xlim(0, 2000)                                    #xrange
plt.ylim(0, 205)                                     #yrange

plt.xticks(np.arange(0, 2001, 500 ))         #ticks
plt.yticks(np.arange(0,201, 40))

plt.legend(frameon=False, fontsize=14)

plt.savefig('HistoWat6.png', dpi = 300, bbox_inches='tight',pad_inches = 0) 
plt.show()
```
